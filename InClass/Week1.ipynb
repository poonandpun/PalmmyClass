<h2>Condition</h2>
<p>เงื่อนไข : สร้างเงื่อนไข => ให้ทำอะไร, ไม่ถูกต้อง ให้ทำอะไร (if - elif -else)</p>

if (condition):
    do something
elif (condition):
    do something
else:
    do something

```python
# A > 10 : print "A more than 10", print "A less than 10"
A = 20
if (A > 10):
    print("A more than 10")
else:
    print("A less than 10")
```

```python
# B > 10 : print "B more than 10",B in 5 - 10 print "B in range", B < 5 : print :"B less than range"
B = 1
if (B > 10):
    print("B more than 10")
elif ((B >= 5) and (B <= 10)):
    print("B in range")
else:
    print("B less than range")
```

```python
B = 10
if (B > 10):
    print("B more than 10")
elif (B in range(5, 11)):
    print("B in range")
else:
    print("B less than range")
```

<h2>Homework solution</h2>

```python
# input student score
def calculateGrade(score):
  if (score >= 80) :
    print("grade = 4")
  elif ((score >= 75) and (score <= 79)):
    print("grade = 3.5")
  elif ((score >= 70) and (score <=74)):
    print("grade = 3")
  elif ((score >= 65) and (score <= 69)):
    print("grade = 2.5")
  elif ((score >= 60) and (score <= 64)):
    print("grade = 2")
  elif ((score >= 55) and (score <= 59)):
    print("grade = 1.5")
  elif ((score >= 50) and (score <= 54)):
    print("grade = 1")
  else:
    print("grade = 0")
```

```python
calculateGrade(75)
calculateGrade(69)
calculateGrade(50)
calculateGrade(65)
```
