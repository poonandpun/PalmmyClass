<h2>For and While loop</h2>

<h3>While loop</h3>
<p>สร้างตัวแปรสำหรับนับ หรือ เช็กเงื่อนไข และมีการอัพเดทตัวนับเสมอ</p>

```python
count = 1
result = []
while count <= 10:
  result.append(count)
  count += 1
print(result)
```

<h3>For loop</h3>
<p>จะต้องมีเป้าหมายการ วนลูปเสมอ</p>

```python
resultForLoop = []
for i in range(1, 11):
  resultForLoop.append(i)
print(resultForLoop)
```

```python
for i in resultForLoop:
  print("Data in list " + str(i))
```
