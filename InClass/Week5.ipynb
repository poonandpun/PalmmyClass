<h2><p>Dataframe  => tools ใช้ในการเอาข้อมูลที่เป็น list, Tuple , dict => ตาราง </p></h2>

```python
import pandas as pd
# install lib (pandas)
```

```python
#weight, hight, age, grade, name,surname
list_weight = [55, 60, 80, 45, 100]
list_hight = [160, 170, 155, 145, 159]
list_age = [12, 13, 12, 12, 13]
list_grade = [4, 3, 2, 4, 2]
list_name = ['A', 'B', 'C', 'D', 'E']
list_surname = ['SurA', 'SurB', 'SurC', 'SurD', 'SurE']
```

<h2>Prepare Data before convert to dataframe<h2>
<p>list => Dict => Dataframe</p>

```python
data = {
    'Name' : list_name,
    'Surname' : list_surname,
    'Weight' : list_weight,
    'Height' : list_hight,
    'Age' : list_age,
    'Grade' : list_grade
}
```

```python
dataTable = pd.DataFrame(data)
```
